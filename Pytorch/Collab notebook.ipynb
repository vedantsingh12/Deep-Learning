```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Each row = one email
# Features: [free_word, money_word, link_present, length]
X = torch.tensor([
    [1, 1, 1, 300],   # spam
    [0, 0, 0, 120],   # not spam
    [1, 0, 1, 250],   # spam
    [0, 1, 0, 100],   # not spam
    [1, 1, 0, 280],   # spam
    [0, 0, 0, 90],    # not spam
], dtype=torch.float32)

y = torch.tensor([
    [1],
    [0],
    [1],
    [0],
    [1],
    [0]
], dtype=torch.float32)
```

```python
class spamClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear=nn.Linear(4,1)
        
    def forward(self,X):
        x= self.linear(X)
        x= torch.sigmoid(X)
        return x
```

```python
model=spamClassifier()
criterion=nn.BCELoss()
optimizer=optim.SGD(model.parameters(),lr=0.1)
```

```python
epochs=200
for epoch in range(epochs):
    y_pred=model(X)
    loss=criterion(y_pred,y)
    
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
if epoch % 20 ==0:
    print(f"epoch{epoch},loss:{loss.item():.4f}")
```

```python
with torch.no_grad():
    preds=model(X)
    preds=(preds>0.5).float()
    
    
    acc=(preds == y).float().mean()
    print("accuracy",acc.item(),"pred:",preds)
```
